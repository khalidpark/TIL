```python
# 제품과 구매에 해당하는 클래스 생성

from datetime import datetime

class User:
    def __init__(self, name, is_admin=False):
        self.name = name
        self.is_admin = is_admin

class Admin(User):
    def __init__(self, name):
        super().__init__(name, is_admin=True)

class Customer(User):
    def __init__(self, name):
        super().__init__(name)
        self.purchases = []

class Seller(User):
    def __init__(self, name):
        super().__init__(name)
        self.products = []


# - 상품 : 이름 , 이미지 , 가격 , 사이즈 , 판매자 , 제조국가
# (Products : Name, Image, Price, Size, Seller, Origin)

class Product:
    def __init__(self, name, image, price, size, seller, origin):
        self.name = name
        self.image = image
        self.price = price
        self.size = size
        self.seller = seller
        self.origin = origin

# - 구매 (+결제) : 상품 , 구매자 , 가격 , 포인트 , 결제방법 , 주소
# (Purchases : Product , Seller, Price, Point, Payment_method, Address)

class Purchase:
    def __init__(self, product, customer, point, method, address):
        self.product = product
        self.customer = customer
        self.purchase_price = product.price
        self.point = point
        self.method = method
        self.address = address
        self.purchase_data = datetime.now()


# - 배송 : 상품, 구매자, 주소 , 배송예상기간
# (Delivery : Address , Delivery_Time)

class Delivery:
    def __init__(self, product, customer, time):
        self.address = product
        self.customer = customer
        self.address = Purchase.address
        self.time = time

# - 커뮤니티 : 상품 , 구매자 (익명) , 판매자
# (Community : Product , Customer , Seller )

class Community:
    def __init__(self, product, customer, seller):
        self.product = product
        self.customer = customer
        self.seller = seller

# - CS :구매자 , 상품
# (CS : Customer , Product)

class CS:
    def __init__(self, product, customer):
        self.customer = customer
        self.product = product
```

```python
# 제품을 구매하는 고객과 제품을 생산하는 공급업체 등 클래스 간의 관계를 소스코드 상에서 구분한다.

from datetime import datetime

class User:
    def __init__(self, name, is_admin=False):
        self.name = name
        self.is_admin = is_admin

class Admin(User):
    def __init__(self, name):
        super().__init__(name, is_admin=True)

class Customer(User):
    def __init__(self, name):
        super().__init__(name)
        self.purchases = []
    def purchase_product(self, product):    # object간의 관계를 위한 추가함수
        purchase = Purchase(product, self)
        self.purchases.append(purchase)

class Vendor(User):
    def __init__(self, name):
        super().__init__(name)
        self.products = []
    def create_product(self, product_name, product_price):  # object간의 관계를 위한 추가함수
        product = Product(product_name, product_price, self)
        self.products.append(product)

# 모델링을 위한 추가 소스코드
class Product:
    def __init__(self, name, price, vendor):
        self.name = name
        self.price = price
        self.vendor = vendor

class Purchase:
    def __init__(self, product, customer):
        self.product = product
        self.customer = customer
        self.purchase_price = product.price
        self.purchase_data = datetime.now()
```
